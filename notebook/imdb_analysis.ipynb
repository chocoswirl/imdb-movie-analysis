```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

```python
df = pd.read_csv("../data/imdb_top_1000.csv")
df.head()
```

```python
df.info
```

```python
df["Released_Year"]=pd.to_numeric(df["Released_Year"],errors="coerce")
```

```python
top_movies = df.sort_values("IMDB_Rating", ascending=False).head(10)
top_movies[["Series_Title", "IMDB_Rating"]]
```

```python
top_movies=df.sort_values("IMDB_Rating", ascending=False).head(10)
top_movies[["Series_Title","IMDB_Rating"]]
```

```python

ratings_by_year=df.groupby("Released_Year")["IMDB_Rating"].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(ratings_by_year.index, ratings_by_year.values)
plt.title("Average IMDb Rating by Year")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
```

```python
genre_counts = df["Genre"].str.split(", ").explode().value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
genre_counts.plot(kind="bar")
plt.title("Top 10 Movie Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.show()
```
